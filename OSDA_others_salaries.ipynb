```python
import pandas as pd, numpy as np
```

```python
data=pd.read_csv("adult.data", header=None)
data.columns=["age", "workclass", "fnlwgt", "education", "education-num",
"marital-status", "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss", "hours-per-week", "native-country", "income"]
data
```

```python
data.info()
```

```python
data = data.replace(" ?", np.NaN)
data.dropna(inplace=True)
data.index=[i for i in range(data.shape[0])]
data=data.replace(" <=50K.", "<=50K")
data=data.replace(" >50K.", ">50K")
data=data.replace(" <=50K", "<=50K")
data=data.replace(" >50K", ">50K")
data
```

```python
[len(np.unique(data[i])) for i in data]
```

```python
def binarize(data):
    bind=pd.DataFrame()
    for i in data:
        if type(data[i].values[0])==str:
            if len(np.unique(data[i]))==2:
                bind[i]=np.zeros(data.shape[0])
                bind.loc[data[i]==np.unique(data[i])[1], [i]]=1
            else:
                for j in np.unique(data[i]):
                    bind[i+"_"+j]=np.zeros(data.shape[0])
                    bind.loc[data[i]==j, [i+"_"+j]]=1
    #             bind[i+'x']=np.zeros(data.shape[0])
    #             bind.loc[data[i]=='x', [i+'x']]=1
        else: bind[i]=data[i]
    return bind
```

```python
beta=binarize(data)
# beta.loc[beta["income"]=="<=50K", ["income"]]=1
# beta.loc[beta["income"]==">50K", ["income"]]=0
beta
```

```python
income=beta["income"]
win=income
```

```python
beta=beta.drop(["income"], axis=1)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
kf = KFold(n_splits=10, shuffle=True)
```

```python
from sklearn.naive_bayes import BernoulliNB
nbc = BernoulliNB()
resultsnbc = cross_val_score(nbc, beta, win, cv=kf)
resultsnbc.mean(), resultsnbc.std()
```

```python
from sklearn.naive_bayes import GaussianNB
nbc = GaussianNB()
resultsnbc = cross_val_score(nbc, beta, win, cv=kf)
resultsnbc.mean(), resultsnbc.std()
```

```python
from sklearn.ensemble import RandomForestClassifier
nbc = RandomForestClassifier()
resultsnbc = cross_val_score(nbc, beta, win, cv=kf)
resultsnbc.mean(), resultsnbc.std()
```

```python
from sklearn.neighbors import KNeighborsClassifier
nbc = KNeighborsClassifier()
resultsnbc = cross_val_score(nbc, beta, win, cv=kf)
resultsnbc.mean(), resultsnbc.std()
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
nbc = LinearDiscriminantAnalysis()
resultsnbc = cross_val_score(nbc, beta, win, cv=kf)
resultsnbc.mean(), resultsnbc.std()
```

```python
from sklearn.linear_model import LogisticRegression
nbc = LogisticRegression()
resultsnbc = cross_val_score(nbc, beta, win, cv=kf)
resultsnbc.mean(), resultsnbc.std()
```

```python
from sklearn.svm import LinearSVC
nbc = LinearSVC()
resultsnbc = cross_val_score(nbc, beta, win, cv=kf)
resultsnbc.mean(), resultsnbc.std()
```

```python
from sklearn.tree import DecisionTreeClassifier
nbc = DecisionTreeClassifier()
resultsnbc = cross_val_score(nbc, beta, win, cv=kf)
resultsnbc.mean(), resultsnbc.std()
```

```python
from sklearn.ensemble import GradientBoostingClassifier
nbc = GradientBoostingClassifier()
resultsnbc = cross_val_score(nbc, beta, win, cv=kf)
resultsnbc.mean(), resultsnbc.std()
```
