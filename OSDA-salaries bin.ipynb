```python
def binarize(data):
    bind=pd.DataFrame()
    for i in data:
        if type(data[i].values[0])==str:
            if len(np.unique(data[i]))==2:
                bind[i]=np.zeros(data.shape[0])
                bind.loc[data[i]==np.unique(data[i])[1], [i]]=1
            else:
                for j in np.unique(data[i]):
                    bind[i+"_"+j]=np.zeros(data.shape[0])
                    bind.loc[data[i]==j, [i+"_"+j]]=1
    #             bind[i+'x']=np.zeros(data.shape[0])
    #             bind.loc[data[i]=='x', [i+'x']]=1
        else: bind[i]=data[i]
    return bind
```

```python
def inter_bin(x, y):
    return x*y
```

```python
def inter_cat(x,y):
    return np.array(list(set((np.append(np.array(x),y)))))
```

```python
def inter_num(x, y):
    return [np.min([x, y]), np.max([x, y])]
```

```python
def inter(x,y, meta):
    res=[]
    for i in range(len(x)):
        if meta[i]=="bin":
            res.append(inter_bin(x[i], y[i]))
        elif meta[i]=="cat":
            res.append(inter_cat(x[i], y[i]))
        elif meta[i]=="num":
            res.append(inter_num(x[i], y[i]))
        else:
            raise Exception(str("Unknown metadata passed: "+meta[i]+" please use one of the following: bin, num, cat"))
    return np.array(res, dtype=object)
```

```python
def isin(x, y, meta):
#     z=inter(x, y, meta)
#     print(z)
    res=True
    for i in range(len(x)):
        if meta[i]=="bin":
            res*=(y[i]>=x[i])
        elif meta[i]=="cat":
            res*=(y[i] in x[i])
        elif meta[i]=="num":
            res*=(x[i][0]<=y[i]<=x[i][1])
        else:
            raise Exception(str("Unknown metadata passed: "+meta[i]+" please use one of the following: bin, num, cat"))
        if not res: return res
    return res
```

```python
def support(x, cl, meta):
    return np.mean([isin(x, i, meta) for i in cl])
```

```python
def conf(x, y, meta, numdiff=0):
#     print(x, y)
    res=[]
    for i in range(len(x)):
        if meta[i]=="bin" or meta[i]=="cat":
            res.append(int(x[i]==y[i]))
        elif meta[i]=="num":
            res.append(int(abs(x[i]-y[i])<=numdiff))
        else:
            raise Exception(str("Unknown metadata passed: "+meta[i]+" please use one of the following: bin, num, cat"))
    return np.mean(res)
```

```python
def classify(x, X_pos, X_neg, min_conf, minsup, aconf, meta, numdif=0):
    cont_pos=0
    sup_pos=0
    npos=0
    minsuppos=np.inf
    maxsuppos=0
    mincontpos=np.inf
    maxcontpos=0
    
    for i in X_pos:
        z=inter(i, x, meta)
        c=conf(i, x, meta, numdif)
#         print(c)
        if c>=min_conf:
#             print("c")
            return 1
        
        
        
        if c>=aconf:
#             print(c)
            sup=support(z, X_pos, meta)
            if sup>=minsup:
                return 1
            cont=support(z, X_neg, meta)
            cont_pos+=cont
            sup_pos+=sup
            npos+=1
            minsuppos=min(minsuppos, sup)
            maxsuppos=max(maxsuppos, sup)
            mincontpos=min(mincontpos, cont)
            maxcontpos=max(maxcontpos, cont)
    #     print("+")
    
    cont_neg=0
    sup_neg=0
    nneg=0
    minsupneg=np.inf
    maxsupneg=0
    mincontneg=np.inf
    maxcontneg=0
    
    for i in X_neg:
        z=inter(i, x, meta)
        c=conf(i, x, meta, numdif)
        if c>=min_conf:
            return 0
        
        if c>=aconf:
#             print(c)
            sup=support(z, X_neg, meta)
            if sup>=minsup:
                return 0
            cont=support(z, X_pos, meta)
            cont_neg+=cont
            sup_neg+=sup
            nneg+=1
            minsupneg=min(minsupneg, sup)
            maxsupneg=max(maxsupneg, sup)
            mincontneg=min(mincontneg, cont)
            maxcontneg=max(maxcontneg, cont)      
#     print("-")
    
    
#     cont_pos/=npos
#     sup_pos/=npos
#     cont_neg/=nneg
#     sup_neg/=nneg
#     print(cont_pos,sup_pos, cont_neg,sup_neg)
#     print(minsupneg,
#             maxsupneg,
#             mincontneg,
#             maxcontneg,"\n" ,minsuppos,
#             maxsuppos,
#             mincontpos,
#             maxcontpos)
#     print(cont_pos,sup_pos, cont_neg,sup_neg)
    if sup_pos!=0 and sup_neg!=0:
        return int(cont_pos/sup_pos<cont_neg/sup_neg)
    elif sup_pos==0 or sup_neg==0 and not sup_pos==sup_neg:
        return int(sup_pos/X_pos.shape[0]>sup_neg/X_neg.shape[0])
    else: return int(cont_pos/X_pos.shape[0]<cont_neg/X_neg.shape[0])
#     return int(maxcontpos<maxcontneg)
#     return int(maxsuppos>maxsupneg)
```

```python
def metrics(y, yn, y_true, yn_true):
    m=dict()
    FN=sum(y^y_true)
    FP=sum(yn^yn_true)
    TP=len(y_true)-FN
    TN=len(yn_true)-FP
    m["FN"]=FN
    m["FP"]=FP
    m["TP"]=TP
    m["TN"]=TN
    m["A"]=(TP+TN)/(FP+FN+TP+TN)
    m["TPR"]=TP/(TP+FN)
    m["TNR"]=TN/(TN+FP)
    m["NPV"]=TN/(TN+FN)
    m["PPV"]=TP/(TP+FP)
    m["FPR"]=FP/(FP+TN)
    m["FDR"]=FP/(FP+TP)
    m["P"]=1/2*(TP/(TP+FP)+TN/(TN+FN))
    m["R"]=1/2*(TP/(TP+FN)+TN/(TN+FP))
    return m
```
